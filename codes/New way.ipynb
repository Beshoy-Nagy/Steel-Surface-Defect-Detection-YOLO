```python
import os
import xml.etree.ElementTree as ET

classes = {
    "crazing": 0,
    "inclusion": 1,
    "patches": 2,
    "pitted_surface": 3,
    "rolled-in_scale": 4,
    "scratches": 5
}

for split in ["train", "val"]:

    label_dir = rf"D:\Projects2\VisionQC\NEU-DET-DETECTION\labels\{split}"

    xml_files = [f for f in os.listdir(label_dir) if f.endswith(".xml")]

    for xml_file in xml_files:

        xml_path = os.path.join(label_dir, xml_file)

        tree = ET.parse(xml_path)
        root = tree.getroot()

        width = int(root.find("size/width").text)
        height = int(root.find("size/height").text)

        txt_path = xml_path.replace(".xml", ".txt")

        with open(txt_path, "w") as f:

            for obj in root.findall("object"):

                cls = obj.find("name").text

                xmin = float(obj.find("bndbox/xmin").text)
                ymin = float(obj.find("bndbox/ymin").text)
                xmax = float(obj.find("bndbox/xmax").text)
                ymax = float(obj.find("bndbox/ymax").text)

                x_center = ((xmin + xmax) / 2) / width
                y_center = ((ymin + ymax) / 2) / height
                box_width = (xmax - xmin) / width
                box_height = (ymax - ymin) / height

                f.write(
                    f"{classes[cls]} "
                    f"{x_center} "
                    f"{y_center} "
                    f"{box_width} "
                    f"{box_height}\n"
                )

print("Done")
```

```python
import os
import shutil

for split in ["train", "val"]:
    base = rf"D:\Projects2\VisionQC\NEU-DET-DETECTION\images\{split}"

    for folder in os.listdir(base):
        folder_path = os.path.join(base, folder)

        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                shutil.move(
                    os.path.join(folder_path, file),
                    os.path.join(base, file)
                )

            os.rmdir(folder_path)

print("Done")
```
