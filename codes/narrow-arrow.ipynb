```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
import yaml

# 1. Define the correct Kaggle paths for your dataset
# (Double-check your Kaggle input folder structure to make sure these match)
data_config = {
    'path': '/kaggle/input/datasets/beshoynagy/new-way/NEU-DET-DETECTION', # The root data folder
    'train': 'images/train',  # path relative to the root 'path' above
    'val': 'images/val',      # path relative to the root 'path' above
    
    # 2. Define your classes (Adjust these names/numbers to match your actual dataset!)
    'nc': 6, 
    'names': ['crazing', 'inclusion', 'patches', 'pitted_surface', 'rolled-in_scale', 'scratches'] 
}

# 3. Save this new configuration to the writable Kaggle working directory
with open('/kaggle/working/custom_data.yaml', 'w') as f:
    yaml.dump(data_config, f, default_flow_style=False)

print("New data.yaml created successfully at /kaggle/working/custom_data.yaml")
```

```python
model.train(
    data="/kaggle/working/custom_data.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)
```
